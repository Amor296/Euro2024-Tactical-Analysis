```python
from statsbombpy import sb
import pandas as pd
import os

# 1. Fetching Euro 2024 Final data
match_id = 3943043 
events = sb.events(match_id=match_id)

# 2. Basic Data Cleaning
events = events.dropna(subset=['location'])

# 3. Match Snapshot: Shooting Stats (Clean Output)
# We use .reset_index() to turn it into a clean DataFrame without dtype info
shots = events[events.type == 'Shot'].groupby('team')['type'].count().reset_index()
shots.columns = ['Team Name', 'Total Shots']

print("--- Match Snapshot ---")
print(shots.to_string(index=False)) # Printing without the index for a cleaner look

# 4. Data Persistence
if not os.path.exists('data'):
    os.makedirs('data')

events.to_csv('data/euro_final_events.csv', index=False)
print("\n✅ Operation successful! Data is ready in the 'data' folder.")
```
